```python
crsp_dsf = conn.raw_sql("""
    select permno, date, ret,prc,shrout
    from crsp.dsf
    where date between '1926-01-01' and '2010-12-31'
    and permno between 10000 and 10060
""")
crsp_dsf.to_csv('crsp_dsf2_sample.csv', index=False)
```

```python
crsp_dsi = conn.raw_sql("""
    select date, vwretd
    from crsp.dsi
    where date between '1926-01-01' and '2010-12-31'
""")
crsp_dsi.to_csv('crsp_dsi.csv', index=False)
```

```python
comp_funda = conn.raw_sql("""
    select gvkey, datadate,fyear, at, sale, xrd, emp, ppegt,ppent,pstkrv,txdb, cogs, invt, sich, consol,indfmt, popsrc, datafmt, dp, dpact,capx,oiadp, oibdp, prcc_f, csho,dltt,ib, ceq, dvp, txdi, txdb, seq,dltt
    from comp.funda
    where datadate between '1926-01-01' and '2010-12-31'
""")
comp_funda.to_csv('comp_funda.csv', index=False)
```

```python
linktable = conn.raw_sql("""
    select gvkey, lpermno, linkdt, linkenddt, linktype
    from crsp.ccmxpf_linktable
    where linktype in ('LU', 'LC')   
    and linkprim in ('P', 'C')                  
""")
linktable.to_csv('linktable.csv', index=False)
```

```python
crsp_msf = conn.raw_sql("""
    select permno, date, prc, shrout
    from crsp.msf
    where date between '1926-01-01' and '2010-12-31'
    and EXTRACT(month FROM date) = 12
""")
crsp_msf.to_csv('crsp_msf.csv', index=False)
```
